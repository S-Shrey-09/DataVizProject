```python
import pandas as pd  #importing pandas library for data extraction and organisation 
import numpy as np  # importing numpy library  for calculations

df= pd.read_csv("./per-capita-renewables.csv") 

#renaming headers
header_names=['Country Name','Code','Year','Renewables per capita (kWh)']
df= pd.read_csv("./per-capita-renewables.csv",header=None,skiprows=1,names=header_names) 

# choosing the specific countries to increase the continental diversity 
lis=['United States','India','Canada','Algeria','Australia','Bangladesh','Brazil','China','Egypt','France','Indonesia','Japan','Luxembourg','Morocco','South Africa','World','New Zealand','Mexico','Denmark','Norway','Peru','Argentina','Poland','Greece','Iran']
df1=df[df['Country Name'].isin(lis)] 
 
lis2=[i for i in range (1999,2020)]
df1=df1[df1['Year'].isin(lis2)]

#dropping a column
df1.drop(['Code'], axis=1, inplace=True)

 # reseting index
df1=df1.reset_index(drop=True)
df1.head(525)
```

```python
df1.describe() # EDA with .describe()
```

```python
# getting Renewables per capita (kWh) values of countries in descending order for years 1999,2004,2009,2014,2019

# year 1999 values
lis3=[1999]
df_1999=df1[df1['Year'].isin(lis3)]
df_1999=df_1999.sort_values(by = 'Renewables per capita (kWh)', axis = 0,ascending= False)
df_1999=df_1999.reset_index(drop=True)
df_1999.head(25)
```
