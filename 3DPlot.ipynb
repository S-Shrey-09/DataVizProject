```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
#importing organised GNI data
GNI = pd.read_csv("https://raw.githubusercontent.com/S-Shrey-09/DataVizProject/main/Selected-organised%20Data%20for%20plotting/GNIdataPlot.csv")

#importing organised renewable data
df_renewable= pd.read_csv("https://raw.githubusercontent.com/S-Shrey-09/DataVizProject/main/Selected-organised%20Data%20for%20plotting/Renewables%20per%20capita%20dataplot.csv")
```

```python
##### from mpl_toolkits.mplot3d import Axes3D
df = pd.DataFrame({"1999":GNI['1999'],"2004":GNI['2004'],"2009":GNI['2009'],"2014":GNI['2014'],"2019":GNI['2019']})
dx, dy = 1, 0.1

# prepare 3d axes
fig = plt.figure(figsize=(9,9))
ax = Axes3D(fig)

# set bar positions on axes
xpos=np.arange(df.shape[0])
ypos=np.arange(df.shape[1])

# set the ticks in the middle of the bars
ax.set_xticks(xpos + dx/2)
ax.set_yticks(ypos +dy/2)

# create X, Y grid 
xpos, ypos = np.meshgrid(xpos, ypos)
xpos = xpos.flatten()
ypos = ypos.flatten()

# set the start of the bar to 0
zpos=np.zeros(df.shape).flatten()
ax.set_zlim3d(0, 70000)

# the bar heights these are the values under the columns 1999-2019
dz = df.values.ravel(order='F')

# ploting the barchart with colours
colours=["red","blue","green","purple","yellow","black","cyan","violet","orange","crimson","pink","brown","gray"]
colour_cycle=[colours[i%5] for i in range(125)]
ax.bar3d(xpos,ypos,zpos,dx,dy,dz,color=colour_cycle)

# Labeling the ticks
ax.w_yaxis.set_ticklabels(["1999","2004","2009","2014","2019"],rotation=45)
ax.w_xaxis.set_ticklabels(GNI['Country Name'],rotation=85,fontsize=12)

# Label the axes
ax.set_ylabel("years")
ax.set_zlabel("GNI")
plt.show()
```
